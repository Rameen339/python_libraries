```python
import matplotlib.pyplot as plt 
import numpy as np   
y = [70, 80, 90]
x = ['C++', 'Python', 'Java']
z=[34,55,47]

plt.xlabel('Language', fontsize=10)
plt.ylabel('Percent', fontsize=10)
plt.title('Bar Graph', fontsize=25)

width=0.3
p=np.arange(len(x))
p1=[j+width for j in p]


plt.bar(x, y,width,color='pink',edgecolor='black',linewidth=2,alpha=0.5,label='Popularity1')
plt.bar(p1, z,width,color='red',edgecolor='black',linewidth=2,alpha=0.5,label='Popularity2')
plt.legend()
plt.xticks(p+width/2,x,rotation=45)
plt.show()
```
